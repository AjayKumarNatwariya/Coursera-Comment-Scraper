```python
from lxml import html
import requests
import pandas as pd
import re
import os
```

```python
dir_path='/content/drive/MyDrive/Untitled folder/'
df=pd.read_csv(dir_path+'u.csv')
```

```python
coursecount=0
for i in range(0,len(df)):
    if str(df.CourseName[i]+'.csv').replace('/',' ') in os.listdir(dir_path+'Courses') or str(df.CourseName[i]+'.csv') in os.listdir(dir_path+'Courses'):
      continue
    user=[]
    dates=[]
    comments=[]
    likes=[]
    ratings=[]
    pag=1
    tot=0
    page=requests.get(df['CourceLink'][i]+"/reviews")
    driver = html.fromstring(page.content)
    numpages=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/nav/ul')
    maxi=0
    num=0
    d=[]
    for com in numpages:
      for z in com.text_content().split( ):
        d.append(z)
    x=filter(str.isdigit,d)
    for com in x:
        try:
            num=int(com)
        except ValueError:
            continue
        if num>maxi:
            maxi=int(num)
    while pag!=maxi+1:
        url=df['CourceLink'][i]+"/reviews?page="+str(pag)
        page=requests.get(url)
        driver = html.fromstring(page.content)
        for j in range(1,26):
            username=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/div[2]/div[{0}]/div[1]/div[1]/p/span'.format(j))
            date=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/div[2]/div[{0}]/div[1]/div[1]/div[2]'.format(j))
            comment=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/div[2]/div[{0}]/div[1]/div[2]/div/div'.format(j))
            like=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/div[2]/div[{0}]/div[2]/button/span/span[2]'.format(j))
            rating=driver.xpath('//*[@id="rendered-content"]/div/div/div[1]/div[4]/div[2]/div[{0}]/div[1]/div[1]/div[1]'.format(j))
            try:
                comment=' '.join([com.text_content() for com in comment])
                name=username[0].text[3:]
                d=date[0].text
                l=''.join(filter(str.isdigit,like[0].text))
                r=[rat.text_content() for rat in rating][0]
            except IndexError:
                continue
            user.append(name)
            dates.append(d)
            comments.append(comment)
            num=0
            try:
                num=int(l)
            except ValueError:
                pass
            likes.append(num)
            ratings.append(int(r.count("F")))
            tot=tot+1
        pag=pag+1
    cdf=pd.DataFrame()
    cdf['UserName']=user
    cdf['Date']=dates
    cdf['Comment']=comments
    cdf['Likes']=likes
    cdf['Rating']=ratings
    if len(cdf)<1:
      continue
    coursecount=coursecount+1
 
    print("Strting Course........",i,coursecount,end="")
    print("Comment Scrapped........",tot,"Pages Scraped.......",pag-1)
    filename=dir_path+"Courses/"+str(df.CourseName[i]).replace('/',' ')+".csv"
    cdf.to_csv(filename)
```
